```python
''' Pull dataset from Kaggle '''

# Install dependencies as needed:
# pip install kaggle
# pip install kagglehub

import pandas as pd
import kagglehub
import shutil
from pathlib import Path

# Dataset files are downloaded to local .kagglehub cache
# Download and get path
download_path = kagglehub.dataset_download("ayaasaad/synthetic-hospital-visits-and-billing-records")

working_dir = Path.cwd()

old_path = Path(download_path)/"Simulated_Clinical_and_Billing_Data_for_Healthcare_Analytics.csv.csv"
data_path = working_dir/"raw_data.csv"

shutil.copy(old_path, data_path)

# Load and preview data
df = pd.read_csv(data_path)
print(df.head())
```
